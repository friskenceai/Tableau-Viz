# Feature Engineering

### changing type of discounted price and actual price to int

#### Discounted price

```python
df['discounted_price']
```

```python
import re
def extract_integer(s):
    # Remove any non-numeric characters
    cleaned_s = re.sub(r'[^\d,]', '', s)
    # Remove commas
    cleaned_s = cleaned_s.replace(',', '')
    # Convert the cleaned string to an integer
    return int(cleaned_s)
```

```python
# Apply the function to the column
df['discounted_price'] = df['discounted_price'].apply(extract_integer)
```

```python
df['discounted_price']
```

#### actual price

```python
df['actual_price']
```

```python
# Apply the function to the column
df['actual_price'] = df['actual_price'].apply(extract_integer)
```

# modifying category column

### Lets divide categires to broader ones

```python
# Function to determine the value of the new column based on presence of keywords
def determine_category(row):
    if 'Mobile' in row['category'] or 'Smartphone' in row['category']:
        return 'Mobile&Accessories'
    else:
        return row['category'].split('|')[0]

# Apply the function to create the new column
df['Broad_Category'] = df.apply(determine_category, axis=1)
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize VADER sentiment intensity analyzer
sid = SentimentIntensityAnalyzer()

# Function to classify reviews as 'good', 'avg', or 'bad' based on sentiment scores
def classify_category(sentiment_score):
    if sentiment_score >= 0.5:
        return 'good'
    elif sentiment_score >= -0.5:
        return 'avg'
    else:
        return 'bad'

# Function to compute the goodness score
def compute_goodness_score(title, review):
    title_sentiment = sid.polarity_scores(title)['compound']
    review_sentiment = sid.polarity_scores(review)['compound']
    
    # Average the sentiment scores of title and review
    goodness_score = (title_sentiment + review_sentiment) / 2
    return goodness_score

# Apply the functions to create new columns
df['review_goodness_score'] = df.apply(lambda x: compute_goodness_score(x['review_title'], x['review_content']), axis=1)
df['review_category'] = df['review_goodness_score'].apply(classify_category)
```
