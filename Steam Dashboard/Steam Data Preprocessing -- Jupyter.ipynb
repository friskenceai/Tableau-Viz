```python
def extract_unique_genres(df):
    unique_genres = set()
    all_genres = set()
    for entry in df['Genres']:
        genres = entry.split(',')
        for genre in genres:
            all_genres.add(genre.strip())
            if len(genres) == 1:
                unique_genres.add(genre.strip())
    # Add single genres that are not part of any combination
    for genre in all_genres:
        if (genre,) not in unique_genres:
            unique_genres.add(genre)
    return list(unique_genres)




# Extract unique combinations of genres
unique_combinations = extract_unique_genres(df)
```

### Modify the Estimated owners column

```python
df1 = pd.read_csv('data_processed.csv')
```

```python
df1[df1['Release date'].isnull()]
```

```python
class_values=df1['Estimated owners'].unique()
```

```python
# Extracting lower and upper bounds from the range strings
lower_bounds = []
upper_bounds = []
for class_val in class_values:
    bounds = class_val.split(' - ')
    lower_bounds.append(int(bounds[0]))
    upper_bounds.append(int(bounds[-1]))

# Calculating the average for each class
averages = [(lower + upper) / 2 for lower, upper in zip(lower_bounds, upper_bounds)]

# Create a DataFrame with class values and their corresponding averages
class_avg_df = pd.DataFrame({'class_value': class_values, 'avg': averages})

# Create a dictionary mapping class values to their averages
class_avg_dict = dict(zip(class_avg_df['class_value'], class_avg_df['avg']))

# Update the existing DataFrame with the calculated averages
df1['Owners_Avg'] = df1['Estimated owners'].map(class_avg_dict)
```
